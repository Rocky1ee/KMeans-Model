```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#对数据进行基本探索
#统计缺失值和其他数字特征
pd.set_option('display.max_columns',None)
df = pd.read_csv('air_data.csv',encoding = 'utf-8')
df.head()
```

```python
df.shape  
```

共有62988个样本，每个样本44个特征

```python
df_des = df.describe().T
df_na = pd.DataFrame([df.isna().sum(),df.isna().sum()/len(df)],index = ['缺失数','缺失率']).T

pd.concat([df_na,df_des],axis=1)
```

工作省份和工作城市的缺失值较多各占3.5%和5%左右。

```python
#缺失数据占比并不大，对于分析问题的影响不大，所以选择丢弃。

#除去空票价和0票价
data = df[df['SUM_YR_1'].notnull() * df['SUM_YR_2'].notnull()]  
index1 = df['SUM_YR_1'] != 0 
index2 = df['SUM_YR_2'] != 0
#除去平均折扣率为0与飞行总公里数同时为0的记录??????
index3 = (df['avg_discount'] == 0 ) & (df['SEG_KM_SUM'] == 0)
data = data[index1|index2|index3]
```

```python
"""
元素数据中的数据太多，根据航空公司客户价值LRFMC模型，选择6个特征：
LOAD_TIME,  FFP_DATE,  FLIGHT_COUNT,  avg_discount,  SEG_KM_SUM,  LAST_TO_END
"""
selected_features = ['LOAD_TIME','FFP_DATE','FLIGHT_COUNT','avg_discount','SEG_KM_SUM',
                   'LAST_TO_END']
data = data[selected_features]
data.head()
```

```python
#构建LRFMC五个特征
#L：入会时长 = 调察结束时间 - 入会时间
data['L'] = pd.to_datetime(data['LOAD_TIME']) - pd.to_datetime(data['FFP_DATE'])

#R:最近乘坐飞机与调察结束的时间间隔 
data['R'] = data['LAST_TO_END']

#F:调查期间内的飞行次数
data['F'] = data['FLIGHT_COUNT']

#M:调查期间内飞行的里程总数
data['M'] = data['SEG_KM_SUM']

#C:观察期间内乘坐舱位折扣系数的平均值
data['C'] = data['avg_discount']

data = data[['L','R','F','M','C']]
```

```python
data.head()
```

```python
#时间戳数值转int数值
data['L'] = data['L'].astype('str').apply(lambda x :int(x[0:4]))
```

```python
data.head()
```

```python
#特征数值差距太大，需要进行标准化处理
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(data)
data_ss = scaler.transform(data)
```

```python
from sklearn.cluster import KMeans
k_cls = KMeans(n_clusters= 5 )
k_cls.fit(data_ss)
k_cls.labels_
```

```python
k_cls.cluster_centers_
```

```python
from collections import Counter

#统计各类别人数
count = Counter(k_cls.labels_)

#创建聚类DataFrame
df_cluster = pd.DataFrame(k_cls.cluster_centers_,columns = 'Center' +data.columns)

df_cluster['聚类个数'] = [count[0],count[1],count[2],count[3],count[4]]

df_cluster['聚类类别'] = ['客户群1','客户群2','客户群3','客户群4','客户群5',]

df_cluster
```

客户群1：入会时间短，离最近一次乘机时间间隔较短，总飞行次数低，总飞行里程低，平均折扣系数低。
从上述信息可以看出该群体属于新客户群体，可以加大宣传，并给予少了优惠政策，保持与客户的一般联系。

从上述信息可以看出该群体属于重要挽留客户群体，应该积极收集客户的最新信息，增强与客户的互动，采取一定的营销手段，延长客户的生命周期。

客户群3：入会时间较短，离最近一次乘机时间间隔久远，总飞行次数低，总飞行里程低，平均折扣率较低。
从上述信息可以看出该群体属于低价值客户，可能是在航班打折时才回乘坐公司的航班。应积极向客户传递折扣信息

客户群4：入会时间久远，离最近一次乘机时间间隔较短，总飞行次数高，总飞行里程长，平均折扣率较高。
